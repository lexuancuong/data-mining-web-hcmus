# Step 2 - Cleaning data

```python
#Làm đầy dữ liệu với những trường NaN trong dữ liệu
user_df['number_of_commets'] = user_df['number_of_commets'].fillna(0)
user_df = user_df[user_df['number_of_tracks'].notna()]
user_df = user_df[user_df['numbert_of_following'].notna()]
user_df = user_df[user_df['number_of_follower'].notna()]
user_df['number_of_likes'] = user_df['number_of_likes'].fillna(0)
track_df.dropna(subset=['name_of_track','link_singer_belong'], inplace=True)
```

**Ngay từ bước xử lí data em đã rất cố gắng để có thể hạn chế tối đa sự trùng lặp trong dữ liệu để hạn chế việc xử lí quá dư thừa ảnh hưởng đến tiến độ. Nên dữ liệu thu được có tỉ lệ trùng lặp gần như bằng 0.**

```python
def process_number_string(num):
    num = str(num)
    num = num.replace(',', '')
    if num[-1]=='M':
        return float(num[:-1])*1000000
    elif num[-1]=='K':
        return float(num[:-1])*1000
    else:
        return float(num)
        
```

```python
user_df.head()
```

**Convert tất cả các trường có trong bảng sang dạng float**

```python
df = user_df
df['number_of_follower'] =df['number_of_follower'].apply(process_number_string)
df['numbert_of_following'] = df['numbert_of_following'].apply(process_number_string)
df['number_of_tracks'] = df['number_of_tracks'].apply(process_number_string)
df['number_of_likes'] = df['number_of_tracks'].apply(process_number_string)
df['number_of_commets'] = df['number_of_commets'].apply(process_number_string)
user_df.info()
```

```python
df = track_df
df['number_of_plays'] =df['number_of_plays'].apply(process_number_string)
df['number_of_repost'] = df['number_of_repost'].apply(process_number_string)
df['number_of_comment'] = df['number_of_comment'].apply(process_number_string)
user_df.info()
```

```python
#Số lượng repost của tất cả các tracks mà User sở hữu
def find_number_of_repost_from_track(link_user):
    return len(track_df[track_df['link_singer_belong'] == link_user])
user_df['number_of_repost_from_tracks'] = user_df['link_of_user_owner'].apply(find_number_of_repost_from_track)
user_df.head()
```

```python
def process_number_string_1(num):
    num = str(num)
    num = num.replace(',', '')
    if num[-1]=='M':
        return float(num[:-1])*1000000
    elif num[-1]=='K':
        return float(num[:-1])*1000
    else:
        return float(num)
def find_number_of_play_from_track(link_user):    
    tmp_rows = track_df[track_df['link_singer_belong'] == link_user]
    return tmp_rows['number_of_plays'].sum()
user_df['number_of_plays_from_all_tracks'] = user_df['link_of_user_owner'].apply(find_number_of_play_from_track)
user_df.head()
```

```python
def find_number_of_comment_from_track(link_user):    
    tmp_rows = track_df[track_df['link_singer_belong'] == link_user]
    return tmp_rows['number_of_comment'].sum()
def get_time_from_number(num):
    return num*3.12
user_df['number_of_comment_from_all_tracks'] = user_df['link_of_user_owner'].apply(find_number_of_play_from_track)
user_df['total_of_all_track_time'] = user_df['number_of_tracks'].apply(get_time_from_number)
user_df.head()
```
